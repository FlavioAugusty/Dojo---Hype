seu primeiro desafio será saber como é a distribuição de idades das pessoas desse dataset. Para isso você deve realizar uma plotagem utilizando histogramas.

````python
# Realize uma plotagem de um histograma para saber a distribuição de idades do dataset

sns.displot(df, x = "Age", binwidth = 3, height=3, facet_kws=dict(margin_titles= True),)
````

Outro recurso muito interessante que temos também é plotar histogramas diferentes para pessoas com doença cardíaca e para pessoas saudáveis.

````python
doentes = df[df['HeartDisease'] == 1]  #seleção da classe 1
doentes = doentes.rename(columns = {'Age':'Age_doente'})   #alteração do nome da coluna

saudaveis = df[df['HeartDisease'] == 0]
saudaveis = saudaveis.rename(columns = {'Age':'Age_saudavel'})


sns.histplot([saudaveis['Age_saudavel'], doentes['Age_doente']])
````

Somente com essa plotagem já fica nítido como que a distribuição de idade é diferente para cada uma das classes. Idades mais avançadas têm uma maior concentração de pessoas com doenças cardíacas, principalmente por volta dos 60 anos.



#### Exercicio 2
Agora, faça o mesmo que feito anteriormente, mas com a variável "MaxHR". isso nos permitirá saber se existe aluma relação entre a frequência cardiaca máxima e a existencia ou não de cardiopatia.

Tente usar os diferentes tipos de histogramas apresentados nos videos.

````python
# Crie um histograma com as duas classes de 'HeartDisease' separadas indicando a distribuição da variável 'MaxHR'
doentes = df[df['HeartDisease'] == 1] #seleção da classe 1
doentes = doentes.rename(columns = {'MaxHR' : 'MaxHR_doente'}) #alteração do nome da coluna

saudaveis = df[df['HeartDisease'] == 0]
saudaveis = saudaveis.rename(columns = {'MaxHR' : 'MaxHR_saudavel'})

sns.histplot([saudaveis['MaxHR_saudavel'], doentes['MaxHR_doente']], kde=True) # Histograma com estimativa de densidade (kde)
plt.show()

sns.histplot([saudaveis['MaxHR_saudavel'], doentes['MaxHR_doente']], element='step') # Histograma com barras em formato de degrau
plt.show()

sns.histplot([saudaveis['MaxHR_saudavel'], doentes['MaxHR_doente']], stat='density', common_norm=False) # Histograma normalizado para área total igual a 1
plt.show()
````

Partindo para plotagens de duas variáveis, temos o gráfico que pode ser dito como um dos mais comuns em DataScience. O gráfico de disperção.

Ele nos mostra a distribuição de um conjunto de dados considerando **duas** variáveis

Escolhendo as variáveis certas é possível realizar uma plotagem e distinguir classes diferentes. O parâmetro "hue" é o responsável por selecionar qual coluna do Dataframe será considerada como alvo

````python
sns.scatterplot(data = df, x = 'Age', y = 'MaxHR', hue = 'HeartDisease')
````

Nesse caso temos um gráfico de disperção com a variável "MaxHR" no eixo y e a variável "Age" no eixo x.

Note que a probabilidade de uma obervação da região superior esquerda ser da classe 0 é muito maior do que na região diagonalmente oposta. Em outras palavras, pessoas com mais idade e com MaxHR mais baixo são mais propensas a serem diagnosticas com cardiopatia.

Um último gráfico que seria legal pontuar devido sua relavância para o DataScience é o **Heatmap**.

Ele nos permite comparar os valores de um dataset entre si a partir das cores.

O Heatmap em especial que nos interessa aqui é o "Heatmap de Correlação".

O conceito de correlação será apresentado mais a frente aqui mesmo no dojo, então não se preocupe se não entender exatamente o que significa, o objetivo aqui é mostrar o gráfico.

De forma resumida, correlação é uma medida que varia de -1 até 1 e diz o quão relacionadas são duas variáveis. Correlações proximas de 0 indicam que uma variável influencia pouco na outra. Já corrrelações mais proximas de 1 ou -1 indicam que existe sim uma relação entre as variáveis.

Esse código nos permite fazer uma matriz de correlção, isto é, uma tabela com todos os valores de correlação entre quaisquer duas variáveis.

```

```

Mas o resultado dele é bastante dificil de visualizar

````python
df.corr(numeric_only=True)
````

Com a função



```
# sns.heatmap(df.corr(), annot = True)
```

O último tipo de gráfico que será abordado nesse módulo é o gráfico de pizza.
É um gráfico muito útil que não temos no Seaborn. Dessa forma, temos que utilizar a função do Matplotlib


```
plt.pie(X)
```
o Link abaixo possui um tutorial de como criar e estilizar os gráficos de pizza pelo Matplotlib

https://medium.com/horadecodar/gr%C3%A1ficos-de-pizza-com-matplotlib-piechart-33ea9760ad87


Um exemplo mais simples que o anterior do seu uso é essa seguinte:

````python
X = [33,52,12,17,42,48] #criação do array de dados
plt.pie(X)  # plotagem do piechart
````

isso gera um gráfico com as classes de tamanho proporcionais no circulo, mas não tem nada escrito no gráfico.

Para adicionar as porcentagens devemos fazer isso:

````python
plt.pie(X, autopct='%.2f%%')
````

o código '%.2f%%' é usado para dizer qual formatação queremos mostrar nas porcentagens. o '2f' ali no meio diz que queremos uma precisão de 2 casas apos a virgula. '3f' seria 3 casas, '1f' seria uma casa e assim por diante.

Podemos adicionar também os nomes das labels. Para isso devemos criar uma lista com os nomes.

````python
nomes = ['classe 1','classe 2','classe 3','classe 4','classe 5','classe 6']

plt.pie(X, autopct='%.2f%%', labels = nomes)
````
