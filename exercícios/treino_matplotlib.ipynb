```python
dados = {
    'X' : [1, 2, 3, 4, 5, 6],
    'Y1' : [120, 110, 130, 145, 118, 125],
    'Y2' : [95,54, 86, 77, 90, 81]
}

base = pd.DataFrame(dados)
```

```python
base.head()
```

```python
# plot
fig, ax = plt.subplots()

ax.plot(base.X, base.Y1, label='Y1')
ax.plot(base.X, base.Y2, label='Y2')

plt.legend()
plt.show()
```

```python
# plot
fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(base.X, base.Y1, label='Y1')
ax.plot(base.X, base.Y2, label='Y2')

plt.legend()
plt.show()
```

```python
# plot
fig, ax = plt.subplots()

ax.plot(base.X, base.Y1, label='Y1')
ax.bar(base.X, base.Y2, label='Y2')

plt.legend()
plt.show()
```

```python
# plot
fig, ax = plt.subplots()

ax.plot(base.X, base.Y1, label='Y1')

ax.annotate("Texto Teste", (1, 135))

plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(base.X, base.Y1, label='Y1')

ax.annotate(base.Y1[0], (base.X[0], base.Y1[0]))

plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(base.X, base.Y1, label='Y1')

for i in range(0, 6):
  ax.annotate(base.Y1[i], (base.X[i], base.Y1[i]))

plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(base.X, base.Y1, label='Y1')

for i in range(0, 6):
  ax.annotate(base.Y1[i], (base.X[i]-0.1, base.Y1[i]+3))

ax.set_yticks([0,160])
ax.yaxis.set_visible(False)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(base.X, base.Y1, label='Y1')

for i in range(0, 6):
  ax.annotate(base.Y1[i], (base.X[i]-0.1, base.Y1[i]+3))

ax.set_yticks([0,160])
ax.yaxis.set_visible(False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(base.X[0:4], base.Y1[0:4], label='Y1')
ax.plot(base.X[3:6], base.Y1[3:6], label='Y2')

for i in range(0, 6):
  ax.annotate(base.Y1[i], (base.X[i]-0.1, base.Y1[i]+3))

ax.set_yticks([0,160])
ax.yaxis.set_visible(False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.show()
```
